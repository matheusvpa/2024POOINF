```java
import java.util.Scanner;

//public class Atv01{
//    public static void main(String[] args) {
        Scanner scanner = new Scanner(System.in);

        System.out.println("Valor total: R$ ");
        double valorTotal = scanner.nextDouble();

        System.out.println("Regular ou especial:");
        String tipoCliente = scanner.next();

        double desconto;
        if (tipoCliente.equalsIgnoreCase("especial")) {
            desconto = 0.10; 
        } 
        else {
            desconto = 0.05; 
        }

        double valorFinal = valorTotal - (valorTotal * desconto);

        System.out.println("Valor final após o desconto: R$" + valorFinal);
        scanner.close();
//    }
//}
```
